```python
import pyautogui
import time
import pandas
import pyperclip

#Entrando no sistema
pyautogui.PAUSE = 0.8

pyautogui.click(x=1030, y=446)
pyautogui.press("esc")
pyautogui.click(x=39, y=1006)
```

```python
#Leitura do banco de dados
dtype_dict= {'chapa': str, 'fitac1': str,'fitac2': str,'fital1': str,'fital2': str}
tabela = pandas.read_excel("Criação cópia.xlsx", sheet_name="Filtragem", dtype=dtype_dict)
print(tabela)

#Looping para digitação dos códigos cadastrados
for linha in tabela.index:
    novo = tabela.loc[linha, "novo"]
    pyautogui.write(novo)
    pyautogui.click(x=865, y=510)
    pyautogui.click(x=869, y=530)
    pyautogui.press("tab")
    pyautogui.press("enter", presses = 3)
    pyautogui.press("enter")
    time.sleep(0.5)
    pyautogui.press("esc")
    pyautogui.click(x=52, y=1000)
```
