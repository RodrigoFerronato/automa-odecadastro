```python
time.sleep(4)
excel_file = "Criação do zero.xlsx"
dtype_dict= {'chapa': str, 'fitac1': str,'fitac2': str,'fital1': str,'fital2': str,'COMP': str,'LARG': str,'ESP': str,'Lado pintado': str,'Lado colado C': str,'Lado colado L': str,'QTDE PINTURA EXTERNA': str,'QTDE PINTURA INTERNA': str,'Valor externo': str,'Valor interno': str}
tabela = pandas.read_excel(excel_file, sheet_name="Filtragem", dtype=dtype_dict, na_values="")
tabela.fillna('', inplace=True)
print(tabela)
pyautogui.press("C")
pyautogui.press("tab")

pyautogui.PAUSE = 0.6
```

```python
for linha in tabela.index:
    novo = tabela.loc[linha, "novo"]
    pyautogui.write(novo)
    pyautogui.press("tab", presses = 4)
    try:
        # Verifica se a imagem está presente na tela com confiança ajustável
        if pyautogui.locateOnScreen('jacadastrado.png', confidence=0.9) is not None:
            pyautogui.press('enter')
            continue  # Pula para a próxima iteração do loop
    except pyautogui.ImageNotFoundException:
        # Se a imagem não for encontrada, continue normalmente
        pass

    pyautogui.write(tabela.loc[linha, "DESCRICAO"])
    pyautogui.press("tab", presses=2)

    pyautogui.write(str(tabela.loc[linha, "chapa"]))               
    pyautogui.press("tab")

    pyautogui.write(str(tabela.loc[linha, "Lado pintado"]))               
    pyautogui.press("tab", presses = 3)

    pyautogui.write(tabela.loc[linha, "uso"])
    pyautogui.press("tab",)

    pyautogui.write(tabela.loc[linha, "Etiqastec"])
    pyautogui.press("tab", presses = 2) 

    pyautogui.write(tabela.loc[linha, "op"])
    pyautogui.press("tab", presses=7)

    pyautogui.write(tabela.loc[linha, "COMP"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "LARG"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "ESP"])
    pyautogui.press("tab", presses = 3)

    pyautogui.write(tabela.loc[linha, "Lado colado C"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "Tipo C"])
    pyautogui.press("tab",)

    pyautogui.write(tabela.loc[linha, "fitac1"])
    pyautogui.press("tab", presses = 4)

    pyautogui.write(tabela.loc[linha, "Lado colado L"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "Tipo L"])
    pyautogui.press("tab",)

    pyautogui.write(tabela.loc[linha, "fital1"])
    pyautogui.press("tab", presses = 4)

    pyautogui.write(tabela.loc[linha, "QTDE PINTURA EXTERNA"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "parametro_externo"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "Valor externo"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "QTDE PINTURA INTERNA"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "parametro_interno"])
    pyautogui.press("tab")

    pyautogui.write(tabela.loc[linha, "Valor interno"])
    pyautogui.press("F2")
    pyautogui.press("tab", presses = 5)
    pyautogui.press("enter")
    
   
   
    
```
